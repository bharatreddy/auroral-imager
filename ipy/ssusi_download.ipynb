```python
import pandas
import datetime
import requests
import bs4
import os
```

```python
# Constants used
baseUrl = "https://ssusi.jhuapl.edu/"
outBaseDir = "../data/"
# Variable inputs
satList = [ "f18", "f17", "f16" ]
inpDate = datetime.datetime( 2015, 4, 9 )
# data type
dataType = "l1b"#"edr-aur"
```

```python
# spc=f16&type=edr-aur&year=2016&Doy=083
inpDoY = inpDate.timetuple().tm_yday
strDoY = str(inpDoY)
if inpDoY < 10:
    strDoY = "00" + str(inpDoY)
if ( inpDoY > 10) & (inpDoY < 100):
    strDoY = "0" + str(inpDoY)

for sat in satList:
    # construct url to get the list of files for the day
    payload = { "spc":sat, "type":dataType,\
               "Doy":strDoY,\
               "year":str(inpDate.year) }
    # get a list of the files from dmsp ssusi website
    # based on the data type and date
    r = requests.get(baseUrl + "data_retriver/",\
                     params=payload, verify=False)
    soup = bs4.BeautifulSoup(r.text, 'html.parser')
    divFList = soup.find("div", {"id": "filelist"})
    hrefList = divFList.find_all(href=True)
    urlList = [ baseUrl + hL["href"] for hL in hrefList ]
    # download the files
    for fUrl in urlList:
        rf = requests.get( fUrl, verify=False )
        currFName = rf.url.split("/")[-1]
        outDir = outBaseDir + dataType + "/" + sat +\
                "/" + inpDate.strftime( "%Y%m%d" ) + "/"
        if not os.path.exists(outDir):
            os.makedirs(outDir)
        with open( outDir + currFName, "wb" ) as ssusiData:
            ssusiData.write( rf.content )
```
